### Importing the required libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
covid_19 = pd.read_csv('districts.csv')
covid_19
```

### Creating dataframe for Gautam Buddha Nagar

```python
gatum_buddha_nagar_covid_19 = covid_19[covid_19["District"].str.contains("Gautam Buddha Nagar")]
confirmed_cases = gatum_buddha_nagar_covid_19["Confirmed"]
recovered_cases = gatum_buddha_nagar_covid_19["Recovered"]
deceased_cases = gatum_buddha_nagar_covid_19["Deceased"]
date = gatum_buddha_nagar_covid_19["Date"]
gatum_buddha_nagar_covid_19
```

## GRAPH OF CHANGE

### CONFIRMED CASES OVER PERIOD OF TIME

```python
gatum_buddha_nagar_covid_19.plot(kind='line', x= 'Date', y='Confirmed')
plt.title("Increase in Number of Cases In Gautam Buddha Nagar District, Since April")
plt.ylabel("Confirmed Cases")
plt.xlabel("Date")
plt.show()


```

### RECOVERD CASES OVER PERIOD OF TIME

```python
gatum_buddha_nagar_covid_19.plot(kind='line', x= 'Date', y='Recovered',color='green')
plt.title("Number of Recoveries In Gautam Buddha Nagar District, Since April")
plt.ylabel("Recovered Cases")
plt.xlabel("Date")
plt.show()
```

### DECEASED CASES OVER PERIOD OF TIME

```python
gatum_buddha_nagar_covid_19.plot(kind='line', x= 'Date', y='Deceased', color='red')
plt.title("Number of Deaths In Gautam Buddha Nagar District, Since April")
plt.ylabel("Deceased Cases")
plt.xlabel("Date")
plt.show()
```

## GRAPH OF VARIATION

### TOTAL CONFIRMED CASES IN DISTRICTS OF UTTAR PRADESH
